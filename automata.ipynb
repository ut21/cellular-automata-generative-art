```python
import numpy as np
import matplotlib.pyplot as plt 
import random
import matplotlib.animation as animation
```

```python
M=5
threshold=29
iteration=5
canvas_size=200
```

```python
plot=np.array([[0 for i in range(canvas_size)] for j in range(canvas_size)])
for i in range(canvas_size):
    for j in range(canvas_size):
        plot[i][j]=random.randint(0,M)
plot_orig=plot.copy()
```

```python
def nbd_check(plot,i,j):
    sum=0
    if plot[i-1][j-1]>plot[i][j]+1:
        sum+=1
    if plot[i-1][j]>plot[i][j]+1:
        sum+=1
    if plot[i-1][j+1]>plot[i][j]+1:
        sum+=1
    if plot[i][j-1]>plot[i][j]+1:
        sum+=1
    if plot[i][j+1]>plot[i][j]+1:
        sum+=1
    if plot[i+1][j-1]>plot[i][j]+1:
        sum+=1
    if plot[i+1][j]>plot[i][j]+1:
        sum+=1
    if plot[i+1][j+1]>plot[i][j]+1:
        sum+=1
    return sum
```

```python
for a in range(iteration):
    for i in range(1, canvas_size-1):
        for j in range(1, canvas_size-1):
            if(nbd_check(plot,i,j)>0):
                plot[i][j]+=1

fig, axes = plt.subplots(1, 2, figsize=(12, 4))

im1=axes[0].imshow(plot_orig, cmap='bone', aspect='equal')
axes[0].set_title("Cellular Automata: Initial State")
axes[0].set_axis_off()
# colorbar1 = plt.colorbar(im1, ax=axes[0])

im2=axes[1].imshow(plot, cmap='bone', aspect='equal')
axes[1].set_title("Cellular Automata: Iteration 5")
axes[1].set_axis_off()
# colorbar1 = plt.colorbar(im2, ax=axes[0])

plt.tight_layout()
plt.show()
```
